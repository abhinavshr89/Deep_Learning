```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman.tail()
plt.plot(batsman['index'], batsman['V Kohli'], marker='o', color='blue', linestyle='-', label='V Kohli')
plt.title("V Kohli's Performance Over the Years")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
```

```python
# Plot Kohli's and Sharma's scores over the years
plt.plot(batsman['index'], batsman['V Kohli'], marker='o', color='green', linestyle='solid', label='V Kohli',linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], marker='o',markersize=10, color='black', linestyle='dashed', label='RG Sharma')

# Adding title and labels
plt.title("Performance of V Kohli and RG Sharma Over the Years")
plt.xlabel("Year")
plt.ylabel("Runs")
plt.legend()
plt.grid(True)
plt.show()
```

### How to fix when a large value is present in the dataset that is way more bigger than the other values in the dataset 

```python
year = [2010, 2011, 2012, 2013, 2014, 2015,2016]  # Replace with actual years
price = [150, 160, 170, 160, 190, 200,1000000]       # Replace with actual prices

# Plotting
plt.plot(year, price, marker='o', color='b', linestyle='-')
plt.title("Price Over Years")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
year = [2010, 2011, 2012, 2013, 2014, 2015,2016]  # Replace with actual years
price = [150, 160, 170, 160, 190, 200,1000000]       # Replace with actual prices

# Plotting
plt.plot(year, price, marker='o', color='b', linestyle='-')
plt.title("Price Over Years")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.xlim(2010,2014) # we can use this to limit our graph on x - axis 
plt.ylim(100,200) # similarly we can use this to limit our graph on y - axis 

plt.show()
```
