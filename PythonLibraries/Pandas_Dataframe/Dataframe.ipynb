# Creating a Dataframe

```python
# using Lists 
data = [
    [1, 'Alice', 85],
    [2, 'Bob', 90],
    [3, 'Charlie', 78]
]

# Creating the DataFrame with column names
df = pd.DataFrame(data, columns=['ID', 'Name', 'Score'])

# Display the DataFrame
df
```

```python
# Using a dictionary
data = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Changin the name of the colomn
marks = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Maths': [85, 78, 92, 63, 88],
    'English': [79, 95, 85, 67, 90],
    'Science': [91, 76, 89, 77, 84]
}

students_df = pd.DataFrame(marks)

# Display the DataFrame
students_df

students_df.rename(columns={'Maths': 'Mathematics'}, inplace=True)
students_df
```

```python
# rowwise adding the subjects and creating a adding a new coloumn in the dataset of Total Marks 
students_df['Total Marks'] = students_df[['Mathematics', 'English', 'Science']].sum(axis=1)
students_df['Average Marks'] = students_df[['Mathematics', 'English', 'Science']].mean(axis=1)
students_df
```

#### Selecting coloumns from a dataframe

```python
math_science = students_df[['Mathematics', 'Science']]
math_science
```
