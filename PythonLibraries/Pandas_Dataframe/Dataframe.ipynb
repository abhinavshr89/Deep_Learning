# Creating a Dataframe

```python
# using Lists 
data = [
    [1, 'Alice', 85],
    [2, 'Bob', 90],
    [3, 'Charlie', 78]
]

# Creating the DataFrame with column names
df = pd.DataFrame(data, columns=['ID', 'Name', 'Score'])

# Display the DataFrame
df
```

```python
# Using a dictionary
data = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```
